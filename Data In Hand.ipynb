# Final Project - Data In Hand

## Hypothesis
I am more likely to like Instagram posts on weekends than during weekdays, suggesting that my personal engagement increases when I have more free time.

## Theoretical Engagement 

Social media engagement is shaped by both how people use the apps and how the platforms show content. On weekends, people usually have more free time, so they scroll more and interact with more posts. The more time someone spends on an app, the more content they see, which increases the chances of liking something.
Because users are more active on weekends, the platform also shows them newer and more relevant posts. So if you use social media more on weekends, it makes sense that you’d end up liking more posts. Overall, engagement tends to follow daily habits and how much free time people have.
Statistical Application

To evaluate the hypothesis, I will analyze Instagram data from my personal data, specifically the liked_posts.json file located in my your_instagram_activity/likes folder. Using that I would compare the data of likes on weekends vs weekdays by using a visual graph/chart

## Planned Data Sources

Liked_posts.json in your_instagram_activity folder

## Reliability
The data is reliably sourced—directly from Instagram’s official data export tool, which records all your likes over the period in question.
The JSON format gives precise timestamp data, which supports accurate day-of-week classification.
Potential Bias / Unreliability
Instagram can remove posts. If a post you liked is later deleted or made private, it may affect outcome

## Limitations
Only likes are analyzed — doesn’t measure how much time I viewed or scrolled through posts.
The data may not reflect deleted or hidden posts
Short time period: it does not reflect long-term behavior.
